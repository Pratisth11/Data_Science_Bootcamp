```python
#Write a Python program to create a Vehicle class with max_speed and mileage instance attributes.

class vehicle:
  def __init__(self, mspeed, mil):
    self.mspeed = mspeed
    self.mil = mil

  def move(self):
    print(" speed is" , self.mspeed ,"mileage is: ",  self.mil)

class speed(vehicle):
  pass

class mileage(vehicle):
  pass

```

```python
s1 = speed('70km/hr', 5)
s1one = speed('80km/hr', 8)
s2 = mileage('5km/hr', 6)
s2two = mileage('6km/hr', 7)

s1.move()
s2.move()
s1one.move()
s2two.move()
```

```python
#Create a Bus class that inherits from the Vehicle class. Give the capacity argument of Bus.seating_capacity() a default value of 50.

class vehicle:
  def __init__(self, mspeed, mil):
    self.mspeed = mspeed
    self.mil = mil

  def move(self):
    print(" speed is" , self.mspeed ,"mileage is: ",  self.mil)
  
  def display_capacity(self,capacity):
    print("capacity" , capacity)

class bus(vehicle):
  def __init__(self, mspeed, mil, capacity):
    capacity = 50
    super().__init__(mspeed,mil)
    super().display_capacity(capacity)
    self.capacity = capacity 
   

  def seating_capacity(self):
    print("capacity is " , self.capacity)
```
