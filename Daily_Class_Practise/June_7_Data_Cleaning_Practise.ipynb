```python
#scatter plot for movie reviews and durations and find out the outliers 

import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('/content/Data Cleaning 1st Example.csv',encoding='latin1')

# Select the columns for the scatter plot
x = df['num_critic_for_reviews']
y = df['duration']

# Plot the scatter plot
plt.scatter(x, y)

# Add labels and title
plt.xlabel('num_critic_for_reviews')
plt.ylabel('duration')
plt.title('Scatter Plot')

# Display the plot
plt.show()
```

```python
#2nd solution
plt.scatter(list(df['num_critic_for_reviews']), list(df['duration']))
plt.xlabel('num_critic_for_reviews')
plt.ylabel('duration')
plt.show()
```
